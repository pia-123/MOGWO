```python
import numpy as np
import pandas as pd
import random
import time
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, KFold
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
from tabulate import tabulate

def GWO_with_CV(objf, lb, ub, dim, SearchAgents_no, Max_iter, X, y):
    Alpha_pos = np.zeros(dim)
    Alpha_score = float("inf")

    Beta_pos = np.zeros(dim)
    Beta_score = float("inf")

    Delta_pos = np.zeros(dim)
    Delta_score = float("inf")

    if not isinstance(lb, list):
        lb = [lb] * dim
    if not isinstance(ub, list):
        ub = [ub] * dim

    Positions = np.zeros((SearchAgents_no, dim))
    for i in range(dim):
        Positions[:, i] = np.random.uniform(0, 1, SearchAgents_no) * (ub[i] - lb[i]) + lb[i]

    Convergence_curve = np.zeros(Max_iter)
    feature_occurrences = np.zeros(dim)

    print("GWO is optimizing \"" + objf.__name__ + "\"")

    timerStart = time.time()
    for l in range(0, Max_iter):
        for i in range(0, SearchAgents_no):
            for j in range(dim):
                Positions[i, j] = np.clip(Positions[i, j], lb[j], ub[j])

            fitness = objf(Positions[i, :], X, y)

            if fitness < Alpha_score:
                Alpha_score = fitness
                Alpha_pos = Positions[i, :].copy()

            if (fitness > Alpha_score and fitness < Beta_score):
                Beta_score = fitness
                Beta_pos = Positions[i, :].copy()

            if (fitness > Alpha_score and fitness > Beta_score and fitness < Delta_score):
                Delta_score = fitness
                Delta_pos = Positions[i, :].copy()

        a = 2 - l * ((2) / Max_iter)

        for i in range(0, SearchAgents_no):
            for j in range(0, dim):
                r1 = random.random()
                r2 = random.random()

                A1 = 2 * a * r1 - a
                C1 = 2 * r2

                D_alpha = abs(C1 * Alpha_pos[j] - Positions[i, j])
                X1 = Alpha_pos[j] - A1 * D_alpha

                r1 = random.random()
                r2 = random.random()

                A2 = 2 * a * r1 - a
                C2 = 2 * r2

                D_beta = abs(C2 * Beta_pos[j] - Positions[i, j])
                X2 = Beta_pos[j] - A2 * D_beta

                r1 = random.random()
                r2 = random.random()

                A3 = 2 * a * r1 - a
                C3 = 2 * r2

                D_delta = abs(C3 * Delta_pos[j] - Positions[i, j])
                X3 = Delta_pos[j] - A3 * D_delta

                Positions[i, j] = (X1 + X2 + X3) / 3

        Convergence_curve[l] = Alpha_score
        feature_occurrences += (Alpha_pos >= 0.4999)

        if (l % 1 == 0):
            print(['At iteration ' + str(l) + ' the best fitness is ' + str(Alpha_score)])
            print('alpha:', np.where(Alpha_pos > 0.5)[0])
            print('Accuracy:', 1 - Alpha_score)  # Display the accuracy

    timerEnd = time.time()
    print('Completed in', (timerEnd - timerStart))

    return Alpha_pos, Convergence_curve, feature_occurrences

def fitness_function(positions, X, y):
    features = np.where(positions >= 0.4999)[0]
    if len(features) == 0:
        return 1  # Returning 1 (which represents 0% accuracy) as no features selected

    train_xf = X.iloc[:, features]

    kf = KFold(n_splits=5, shuffle=True, random_state=42)
    accuracies = []
    for train_index, test_index in kf.split(train_xf):
        X_train, X_val = train_xf.iloc[train_index], train_xf.iloc[test_index]
        y_train, y_val = y.iloc[train_index], y.iloc[test_index]

        dt_classifier = DecisionTreeClassifier()
        dt_classifier.fit(X_train, y_train)
        accuracy = dt_classifier.score(X_val, y_val)
        accuracies.append(accuracy)

    avg_accuracy = np.mean(accuracies)
    return 1 - avg_accuracy  # Minimize error (1 - accuracy)

def calculate_correlations(X, y):
    correlations = X.apply(lambda x: x.corr(y))
    return correlations

# Load the data
data_df = pd.read_excel('/content/Copy of CTG_new(1).xlsx')
num_features = 21
train_data, test_data = train_test_split(data_df, test_size=0.3, random_state=42)

train_x, test_x = train_data.iloc[:, :num_features], test_data.iloc[:, :num_features]
train_y, test_y = train_data.iloc[:, -1], test_data.iloc[:, -1]

# Feature selection using GWO
Max_iter = 30
result = GWO_with_CV(fitness_function, 0, 1, num_features, 50, Max_iter, train_x, train_y)
best_solution = result[0]
feature_occurrences = result[2]

# Plot the convergence curve
plt.plot(result[1])
plt.xlabel('Iterations')
plt.ylabel('Fitness')
plt.title('Convergence Curve')
plt.show()

# Calculate correlations with the target
correlations = calculate_correlations(train_x, train_y)

# Create DataFrame for ratings and rankings
features_df = pd.DataFrame({
    'Feature': train_x.columns,
    'Correlation': correlations,
    'Occurrences': feature_occurrences
})

# Normalize the ratings to a scale of 1-5
features_df['Rating-1'] = (features_df['Correlation'].rank(method='max', ascending=False) / features_df['Correlation'].rank(method='max', ascending=False).max()) * 4 + 1
features_df['Rating-2'] = (features_df['Occurrences'].rank(method='max', ascending=False) / features_df['Occurrences'].rank(method='max', ascending=False).max()) * 4 + 1

# Calculate average rating
features_df['Average_Rating'] = features_df[['Rating-1', 'Rating-2']].mean(axis=1)

# Rank the features based on average rating
features_df['Rank'] = features_df['Average_Rating'].rank(method='min', ascending=True)

# Select top 5 features
top_features = features_df.nsmallest(5, 'Rank')
print(top_features)

# Save the rankings to a table (e.g., CSV or display as a table)
print(tabulate(top_features, headers='keys', tablefmt='pretty'))

# You can save the results to a CSV file if needed
top_features.to_csv('top_features.csv', index=False)
```
